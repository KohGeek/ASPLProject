```python
from ultralytics import YOLO
from ray import tune

def tune_model(prefix, pt_model, data, space):
    name = prefix + '_' + pt_model.split('.')[0] + '_' + data.split('.')[0]

    task = pt_model.split('-')[1].split('.')[0]
    if task == 'seg':
        task = 'segment'
    elif task == 'obb':
        task = 'OBB'
    else:
        raise ValueError('Invalid task type. Must be seg or obb')

    model = YOLO(pt_model)
    model.task = task
    result_grid = model.tune(data=data, use_ray=True, space=space, grace_period=8, batch_size=-1, epoch=40, name=name)
   
    return result_grid

# models = {'yolov8s-seg.pt': [640,0.3,0],'yolov9c-seg.pt': [640,0.3,0], 'yolov8s-obb.pt': [800,0.2,0.25]}
models = ['yolov8s-seg.pt', 'yolov9c-seg.pt', 'yolov8s-obb.pt']

datasets = ['cgi.yaml', 'human.yaml', 'combined.yaml']

# space = {
#     'lr0': tune.uniform(1e-5, 1e-1),
#     'lrf': tune.uniform(0.01, 1),
#     'momentum': tune.uniform(0.6, 0.98),
#     'weight_decay': tune.uniform(0.0, 0.001),
#     'warmup_epochs': tune.randint(0, 5),
#     'warmup_momentum': tune.uniform(0.0, 0.95),
#     'box': tune.uniform(0.02, 8),
#     'cls': tune.uniform(0.2, 4.0),
#     'hsv_h': tune.uniform(0.0, 0.1),
#     'hsv_s': tune.uniform(0.0, 0.9),
#     'hsv_v': tune.uniform(0.0, 0.9),
#     'degrees': tune.uniform(-180.0, 180.0),
#     'translate': tune.uniform(0.0, 0.9),
#     'scale': tune.uniform(0.0, 0.9),
#     'perspective': tune.uniform(0, 0.001),
#     'mosaic': tune.uniform(0, 1),
#     'mixup': tune.uniform(0, 1),
#     'copy_paste': tune.uniform(0, 1),
#     'dropout': tune.uniform(0, 0.5)
# }

space = {
    'lr0': tune.uniform(1e-5, 1e-1)
}

for pt in models:
    for dataset in datasets:
        result = tune_model('tuner', pt, dataset, space)
```
